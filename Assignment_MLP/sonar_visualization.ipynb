```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('sonar.csv')
```

```python
data.drop(columns=['Unnamed: 0', 'J', 'I', 'H', 'G'],inplace= True)
```

```python
data.info()
```

```python
print(data)
```

```python
plt.scatter(data.A,data.B,c=data.Class)
plt.colorbar()
plt.xlabel("column A")
plt.ylabel("column B")
plt.title("data from column A and B")
plt.show()
```
